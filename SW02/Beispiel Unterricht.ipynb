# Computer Algebra mit sympy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(t):
    return t**2
def y(t):
    return t*(1-t)+2/(1+(50*(t-1/2))**2)
```

```python
tp = np.linspace(0,1,4000)
```

```python
plt.plot(tp,x(tp))
plt.plot(tp,y(tp))
```

```python
plt.plot(tp,x(tp)-y(tp))
plt.grid()
```

```python
plt.plot(tp,-x(tp)+y(tp))
```

```python
plt.plot(tp,np.abs(x(tp)-y(tp)))
```

```python
tp[np.argmax(np.abs(x(tp)-y(tp)))]
```

Wie man es **nicht** machen sollte:

```python
i = 0
d = np.abs(x(tp)-y(tp))
imax = 0
dmax = d[0]
for i in range(tp.shape[0]):
    if d[i] >= dmax:
        imax = i
        dmax = d[i]
```

```python
import sympy as sp
```

```python
t = sp.symbols('t')
```

```python
x(t)
```

```python
x(t)-y(t)
```

```python
sp.solve(x(t)-y(t),t)
```

```python
textr=sp.solve(sp.diff(-x(t)+y(t)),t)
```

```python
textr
```

```python
xy2 = sp.diff(-x(t)+y(t),t,2)
```

```python
g = sp.lambdify(t,xy2)
```

```python
g(np.linspace(0,1,400))
```

```python
for ti in textr[:3]:
    print(xy2.subs(t,ti))
```
